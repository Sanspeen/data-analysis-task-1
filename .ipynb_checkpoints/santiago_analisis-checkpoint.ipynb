# Manipulacion de datos

1. Elmine columnas que contengan datos duplicados

2. Para las columnas categóricas, unifique los valores con significados duplicados.

```python
ventas.lleva_libros = ventas.lleva_libros.fillna(0)
ventas.lleva_tabaco = ventas.lleva_tabaco.fillna(0)
ventas.lleva_alcohol = ventas.lleva_alcohol.fillna(0)
ventas.lleva_bolsa_plastica = ventas.lleva_bolsa_plastica.fillna(0)
ventas.franquicia_tdc = ventas.franquicia_tdc.replace(["Mastercard", "American Express"], ["MCD", "AMEX"])
ventas.medio_pago = ventas.medio_pago.replace(["Tarjeta", "Efectivo"], ["TDC", "Cash"])
ventas.moneda = ventas.moneda.replace(["euros", "dolares americanos", "pesos colombianos"], ["EUR", "USD", "COP"])
ventas.tipo_viaje = ventas.tipo_viaje.replace("Local", "Nacional")
ventas.tipo_transito = ventas.tipo_transito.replace(["Arrival", "Departure"], ["Llegada", "Salida"])
 
isCarryingBooks = ventas["lleva_libros"].unique().tolist()
print(f"Lleva o no libros: {isCarryingBooks}")

isCarryingTabaco = ventas["lleva_tabaco"].unique().tolist()
print(f"Lleva o no tabaco: {isCarryingTabaco}")
 
isCarryingAlcohol = ventas["lleva_alcohol"].unique().tolist()
print(f"Lleva o no alcohol: {isCarryingAlcohol}")

isCarryingBag = ventas["lleva_bolsa_plastica"].unique().tolist()
print(f"Lleva o no bolsa: {isCarryingBag}")

franchiseTypes = ventas["franquicia_tdc"].unique().tolist()
print(f"Tipos de franquicia tdc: {franchiseTypes}")

paymentTypes = ventas["medio_pago"].unique().tolist()
print(f"Tipos de medios de pago: {paymentTypes}")

cashTypes = ventas["moneda"].unique().tolist()
print(f"Tipos de moneda: {cashTypes}")

travelTypes = ventas["tipo_viaje"].unique().tolist()
print(f"Tipos de viaje: {travelTypes}")
```

# Agrupacion de datos
1. ¿Cuánto fue el total de ventas de cada tienda?

```python
pd.options.display.float_format = '{:.1f}'.format
print("Agrupamos las ventas por el identificador de cada una de las tiendas y tambien lo dividimos por el tipo de divisa que recibio la creandonos asi montos totales de ventas con cada una de las divisas por moneda.")
shop_sells = ventas.groupby(by = ["id_tienda", "moneda"]).monto.sum()
shop_sells_dataframe = pd.DataFrame(shop_sells)
shop_sells_dataframe
```

2. ¿Cuánto fue el total de ventas en pesos colombianos de cada tienda? Utilice el valor actual de la divisa para la conversión.

```python
sells_per_shop = []
EURO_TO_PESOS = 4432.72
DOLAR_TO_PESOS = 4028.85
for i in range(100, 1100, 100):
    total_sells = 0
    converted_euros_sells = shop_sells[i]["EUR"] * EURO_TO_PESOS  
    converted_dolar_sells = shop_sells[i]["USD"] * DOLAR_TO_PESOS
    total_sells = total_sells + converted_euros_sells + converted_dolar_sells + shop_sells[i]["COP"]
    sells_per_shop.append(total_sells)
```

```python
id_stores = ventas.id_tienda.unique().tolist()
sells_per_shop_dataframe = pd.DataFrame({"id_tienda":pd.Series(id_stores).sort_values().tolist(), "total_vendido": sells_per_shop})
sells_per_shop_dataframe
```
